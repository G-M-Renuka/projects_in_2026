# Encode Categorical variables(Location,Condition)

```python
print("Columns in DataFrame:")
print(df.columns.tolist())


df = pd.get_dummies(df, columns=['Neighborhood', 'Condition1'], drop_first=True)
print("DataFrame after one-hot encoding 'Neighborhood' and 'Condition1':")
print(df)
```

# **Split data into training and testing sets**



```python
from sklearn.model_selection import train_test_split

X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

# Identify categorical columns that are still of object type
categorical_cols = X.select_dtypes(include='object').columns

# Apply one-hot encoding to remaining categorical columns
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2
)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

# **Train Regression Model**

```python
from sklearn.linear_model import LinearRegression

# Check for NaNs in X_train before training
missing_in_X_train = X_train.isnull().sum()
missing_in_X_train = missing_in_X_train[missing_in_X_train > 0]

if not missing_in_X_train.empty:
    print("NaNs found in X_train before training:")
    print(missing_in_X_train)
    # Drop columns with NaNs for now, or consider a more robust imputation strategy
    X_train_cleaned = X_train.dropna(axis=1) # Drop columns with NaNs
    # Align X_test with cleaned X_train columns
    X_test_cleaned = X_test[X_train_cleaned.columns]
    print("Dropped columns with NaNs from X_train and X_test.")
else:
    print("No NaNs found in X_train.")
    X_train_cleaned = X_train
    X_test_cleaned = X_test

model = LinearRegression()
model.fit(X_train_cleaned, y_train)
```

#  8.**Identifying** **Price**-**Influencing** **Factors**

**Graph** **6**: Feature Importance(Random Forest)

Purpose:



*  Identify the most influential factors affecting property price

Graph:
 Bar chart of Feature Importance



```python
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt


# Train model
model = RandomForestRegressor(random_state=42)
model.fit(X, y)

# Feature importance
importances = model.feature_importances_
feature_names = X.columns # Define feature_names

# Bar chart
plt.figure(figsize=(10, 6)) # Add a figure to make the plot readable
plt.bar(feature_names, importances)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance (Random Forest)")
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
